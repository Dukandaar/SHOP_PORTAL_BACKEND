```python
# Load Movie Titles
movies_url = "https://files.grouplens.org/datasets/movielens/ml-100k/u.item"
movies_df = pd.read_csv(movies_url, sep="|", encoding="latin-1", header=None, usecols=[0, 1], names=["movie_id", "title"])
movies_df["movie_id"] -= 1  # Zero-based index
movies_df = movies_df.sort_values("movie_id").reset_index(drop=True)
movies_df.head()
```

```python
# Load Sentence-BERT for Movie Titles (Ensure 768-D output)
sbert_model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")  # 768-D output
movie_embeddings = sbert_model.encode(movies_df["title"].tolist(), convert_to_tensor=True)
```

```python
# Check BERT output size
embedding_dim = movie_embeddings.shape[1]  # 768
print(embedding_dim)
# Create movie embedding layer
movie_embedding_layer = nn.Embedding.from_pretrained(movie_embeddings, freeze=True).to(device)
```

```python
# Split data into train & test
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Convert to PyTorch tensors and move to device
train_users = torch.tensor(train_data["user_id"].values, dtype=torch.long).to(device)
train_movies = torch.tensor(train_data["movie_id"].values, dtype=torch.long).to(device)
train_ratings = torch.tensor(train_data["rating"].values, dtype=torch.float32, device=device) #/ 5.0  # Normalize

test_users = torch.tensor(test_data["user_id"].values, dtype=torch.long).to(device)
test_movies = torch.tensor(test_data["movie_id"].values, dtype=torch.long).to(device)
test_ratings = torch.tensor(test_data["rating"].values, dtype=torch.float32, device=device) #/ 5.0  # Normalize
```

```python
# Define GNN Layer
class GNNLayer(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(GNNLayer, self).__init__()
        self.conv = SAGEConv(in_channels, out_channels)

    def forward(self, x, edge_index):
        return self.conv(x, edge_index).relu()
```

```python
class HybridBERT_GNN_NCF(nn.Module):
    def __init__(self, num_users, embedding_dim=768, dropout_rate=0.3):
        super(HybridBERT_GNN_NCF, self).__init__()

        self.embedding_dim = embedding_dim
        self.gnn = GNNLayer(embedding_dim, embedding_dim)

        # GNN-based User Embeddings (Match BERT size)
        self.user_embedding = nn.Embedding(num_users, embedding_dim)

        # Pretrained BERT Movie Embeddings
        self.movie_embedding = movie_embedding_layer

        # Fully Connected Layers
        self.fc1 = nn.Linear(embedding_dim * 2, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 1)

        # Activation & Regularization
        self.leaky_relu = nn.LeakyReLU(0.1)  # Prevents dying neurons
        self.dropout = nn.Dropout(dropout_rate)
        self.batch_norm1 = nn.BatchNorm1d(512)
        self.batch_norm2 = nn.BatchNorm1d(256)

    def forward(self, user, movie, edge_index):
        # User embeddings from GNN
        user_emb = self.user_embedding(user)

        # Movie embeddings from pre-trained BERT
        movie_emb = self.movie_embedding(movie)

        # Apply GNN for user embeddings
        node_features = torch.cat([self.user_embedding.weight, self.movie_embedding.weight], dim=0)
        x_gnn = self.gnn(node_features, edge_index)

        user_gnn = x_gnn[user]
        movie_gnn = movie_emb  # Directly use BERT embeddings

        # Concatenate user & movie embeddings
        x_final = torch.cat([user_gnn, movie_gnn], dim=1)

        # Fully connected layers with improved flow
        x_final = self.batch_norm1(self.fc1(x_final))  # Normalize first
        x_final = self.leaky_relu(x_final)
        x_final = self.dropout(x_final)

        x_final = self.batch_norm2(self.fc2(x_final))
        x_final = self.leaky_relu(x_final)
        x_final = self.dropout(x_final)

        x_final = self.leaky_relu(self.fc3(x_final))

        return self.fc4(x_final).squeeze()  # Output single prediction
```

```python
# Initialize Model
model = HybridBERT_GNN_NCF(num_users, embedding_dim).to(device)

# Loss and Optimizer
criterion = nn.SmoothL1Loss()
optimizer = optim.Adam(model.parameters(), lr=0.0005, weight_decay=1e-5)
```

```python
# Train Model
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    predictions = model(train_users, train_movies, data.edge_index)
    loss = criterion(predictions, train_ratings)
    loss.backward()
    optimizer.step()

    if epoch % 5 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.6f}")
```

```python
# Evaluate Model
model.eval()
with torch.no_grad():
    test_predictions = model(test_users, test_movies, data.edge_index)
    test_loss = criterion(test_predictions, test_ratings)

print(f"\nTest Loss: {test_loss.item():.4f}")
```

```python
# Convert movie_id to zero-based index
movies_df["movie_id"] -= 1
movie_dict = dict(zip(movies_df["movie_id"], movies_df["title"]))

# Function to recommend movies with titles
def recommend_movies(user_id, top_n=5):
    model.eval()
    with torch.no_grad():
        embeddings = model(data)
        user_embedding = embeddings[user_id]
        movie_embeddings = embeddings[num_users:]  # Movies start after users
        similarities = torch.matmul(movie_embeddings, user_embedding)
        recommended_movie_ids = similarities.argsort(descending=True)[:top_n]

        # Convert movie IDs to titles
        recommended_movie_titles = [movie_dict[movie_id.item()] for movie_id in recommended_movie_ids]
        return recommended_movie_titles
```

```python
# movie_recommender.py

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from tqdm import tqdm
from sklearn.manifold import TSNE

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1) LOAD & PREPROCESS
ratings = pd.read_csv(
    "https://files.grouplens.org/datasets/movielens/ml-100k/u.data",
    sep="\t", names=["user_id","movie_id","rating","timestamp"]
)
ratings["user_id"] -= 1
ratings["movie_id"] -= 1
num_users = ratings.user_id.nunique()
num_movies = ratings.movie_id.nunique()

movies = pd.read_csv(
    "https://files.grouplens.org/datasets/movielens/ml-100k/u.item",
    sep="|", encoding="latin-1", header=None, usecols=[0,1],
    names=["movie_id","title"]
)
movies["movie_id"] -= 1
movies = movies.sort_values("movie_id").reset_index(drop=True)

# Normalize ratings for edge weights
ratings["r_norm"] = ratings.rating / 5.0

# 2) GRAPH CONSTRUCTION
# edges: user -> movie (offset), make bidirectional
e1 = torch.tensor(
    [[u, num_users+m] for u,m in zip(ratings.user_id, ratings.movie_id)],
    dtype=torch.long
).t()
edge_index = torch.cat([e1, e1[[1,0]]], dim=1)  # bidirectional

w = torch.tensor(ratings.r_norm.values, dtype=torch.float32)
edge_weight = torch.cat([w, w], dim=0).to(device)

# 3) NODE FEATURES
# Movie features from SBERT
sbert = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")
movie_embs = sbert.encode(movies.title.tolist(), convert_to_tensor=True)
movie_embs = movie_embs.to(device)
emb_dim = movie_embs.size(1)

# User features: learnable
user_embs = nn.Parameter(torch.randn(num_users, emb_dim), requires_grad=True)

# Combine into X
X = torch.cat([user_embs, movie_embs], dim=0)

data = Data(edge_index=edge_index.to(device), edge_attr=edge_weight)
data.x = X

# 4) MODEL DEFINITION
class GCNBlock(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.conv1 = GCNConv(dim, dim)
        self.conv2 = GCNConv(dim, dim)
        self.bn1 = nn.BatchNorm1d(dim)
        self.bn2 = nn.BatchNorm1d(dim)
    def forward(self, x, edge_index, edge_weight):
        x = F.relu(self.bn1(self.conv1(x, edge_index, edge_weight)))
        x = F.relu(self.bn2(self.conv2(x, edge_index, edge_weight)))
        return x

class HybridRecommender(nn.Module):
    def __init__(self, num_users, emb_dim):
        super().__init__()
        self.gcn = GCNBlock(emb_dim)
        self.user_emb = user_embs  # Parameter from above
        self.movie_emb_layer = nn.Embedding.from_pretrained(movie_embs, freeze=True)
        # MLP head
        self.fc1 = nn.Linear(emb_dim*3, 512)
        self.fc2 = nn.Linear(512,256)
        self.fc3 = nn.Linear(256,128)
        self.fc4 = nn.Linear(128,1)
        self.dropout = nn.Dropout(0.5)
        self.bn1 = nn.BatchNorm1d(512)
        self.bn2 = nn.BatchNorm1d(256)
    def forward(self, users, movies, edge_index, edge_weight):
        # GCN propagate
        x = self.gcn(data.x, edge_index, edge_weight)
        # pick embeddings
        u_gcn = x[users]
        m_gcn = x[num_users+movies]
        m_bert = self.movie_emb_layer(movies)
        # concat
        h = torch.cat([u_gcn, m_gcn, m_bert], dim=1)
        h = F.relu(self.bn1(self.fc1(h))); h = self.dropout(h)
        h = F.relu(self.bn2(self.fc2(h))); h = self.dropout(h)
        h = F.relu(self.fc3(h))
        return self.fc4(h).squeeze()

model = HybridRecommender(num_users, emb_dim).to(device)
opt = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
sched = optim.lr_scheduler.ReduceLROnPlateau(opt, patience=3, factor=0.5)

# 5) TRAIN/VALID SPLIT
train_df, val_df = train_test_split(ratings, test_size=0.2, random_state=42)
def make_batch(df):
    return (
      torch.tensor(df.user_id.values, dtype=torch.long).to(device),
      torch.tensor(df.movie_id.values, dtype=torch.long).to(device),
      torch.tensor(df.r_norm.values, dtype=torch.float32).to(device)
    )
u_tr, m_tr, y_tr = make_batch(train_df)
u_val, m_val, y_val = make_batch(val_df)

# 6) TRAINING LOOP
epochs = 30
train_losses, val_losses = [], []
for ep in range(1, epochs+1):
    model.train()
    opt.zero_grad()
    preds = model(u_tr, m_tr, data.edge_index, data.edge_attr)
    loss = F.mse_loss(preds, y_tr)
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), 2.0)
    opt.step()
    # val
    model.eval()
    with torch.no_grad():
        vpred = model(u_val, m_val, data.edge_index, data.edge_attr)
        vloss = F.mse_loss(vpred, y_val)
    sched.step(vloss)
    train_losses.append(loss.item()); val_losses.append(vloss.item())
    print(f"Epoch {ep:02d} | Train MSE {loss:.4f} | Val MSE {vloss:.4f}")

# 7) EVALUATION
model.eval()
with torch.no_grad():
    all_preds = model(u_val, m_val, data.edge_index, data.edge_attr).cpu().numpy()
    true = y_val.cpu().numpy()
rmse = np.sqrt(mean_squared_error(true, all_preds))
mae  = mean_absolute_error(true, all_preds)
print(f"\nValidation RMSE: {rmse:.4f} | MAE: {mae:.4f}")

# Precision@K, Recall@K, NDCG@K
def precision_recall_ndcg(k):
    precisions, recalls, ndcgs = [],[],[]
    grouped = val_df.groupby("user_id")
    for uid, group in grouped:
        seen = set(group.movie_id)
        # candidates
        candidates = [m for m in range(num_movies) if m not in seen]
        u_repeat = torch.tensor([uid]*len(candidates),dtype=torch.long).to(device)
        m_cand = torch.tensor(candidates,dtype=torch.long).to(device)
        with torch.no_grad():
            scores = model(u_repeat, m_cand, data.edge_index, data.edge_attr).cpu().numpy()
        topk = np.argsort(-scores)[:k]
        recs = [candidates[i] for i in topk]
        # relevant = those in val_df with rating>=4
        rel = set(group[group.rating>=4].movie_id)
        hit = len(set(recs)&rel)
        precisions.append(hit/k)
        recalls.append(hit/max(len(rel),1))
        # NDCG
        dcg = sum((1/(np.log2(i+2))) for i,m in enumerate(recs) if m in rel)
        idcg= sum(1/(np.log2(i+2)) for i in range(min(len(rel),k)))
        ndcgs.append(dcg/(idcg+1e-8))
    return np.mean(precisions), np.mean(recalls), np.mean(ndcgs)

for k in [5,10,20]:
    p,r,n = precision_recall_ndcg(k)
    print(f"@{k} Precision: {p:.4f} Recall: {r:.4f} NDCG: {n:.4f}")

# 8) LOSS CURVES
plt.figure(figsize=(6,4))
plt.plot(train_losses,label="Train MSE")
plt.plot(val_losses,label="Val MSE")
plt.xlabel("Epoch"); plt.ylabel("MSE"); plt.legend(); plt.title("Training Curves")
plt.show()

# 9) EMBEDDING VISUALIZATION (t-SNE)
# sample 2000 nodes
idx = np.random.choice(num_users+num_movies,2000,replace=False)
embs = model.gcn(data.x, data.edge_index, data.edge_attr).detach().cpu().numpy()[idx]
tsne = TSNE(n_components=2, random_state=42).fit_transform(embs)
plt.figure(figsize=(6,6))
plt.scatter(tsne[:,0], tsne[:,1], s=5, c=['r' if i<num_users else 'b' for i in idx])
plt.title("t-SNE of GCN Node Embeddings\n(Red=Users, Blue=Movies)"); plt.show()

# 10) PLOT SAMPLE GRAPH
G = to_networkx(data, to_undirected=True)
plt.figure(figsize=(5,5))
nx.draw(G.subgraph(list(range(50))+list(range(num_users,num_users+50))),
        node_size=20, edge_color='gray')
plt.title("Sample Subgraph"); plt.show()

# 11) RECOMMENDATION FUNCTION
def recommend(user_id, top_k=10):
    seen = set(ratings[ratings.user_id==user_id].movie_id)
    candidates = [m for m in range(num_movies) if m not in seen]
    u_rep = torch.tensor([user_id]*len(candidates),dtype=torch.long).to(device)
    m_rep = torch.tensor(candidates,dtype=torch.long).to(device)
    with torch.no_grad():
        scores = model(u_rep,m_rep,data.edge_index,data.edge_attr).cpu().numpy()
    top = np.argsort(-scores)[:top_k]
    return movies.iloc[[candidates[i] for i in top]][["title"]]

# Example:
print("\nTop 10 for user 0:\n", recommend(0,10))
```
